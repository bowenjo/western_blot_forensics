```python
import numpy as np 
import matplotlib.pyplot as plt

import analysis.HOG_analyze as analyzer
import params.config_synthetic as config
%matplotlib notebook
```

```python
save_folder = 'Projects/test_synthetic/saves/'
save_files = ['blur', 'noise', 'contrast', 'scale', 'rotate', 'reflect']
x_labels = ['Kernel size', 'Std dev', 'Contrast gain', 'Scale factor', 'Degrees', 'Axis']
```

```python
fig, ax = plt.subplots(2,3, figsize=(13,9))

count = 0
for i in range(2):
    for j in range(3):
        name, args, labels = config.get_args(save_files[count])
        pre_load = np.load(save_folder + save_files[count] +'.npy').item()
        analyzer.draw_auc(ax[i,j],  pre_load["AUC_Dists"], pre_load["AUC_Ratios"],
                         labels, name, x_labels[count])
        count+=1 
fig.tight_layout()
```
